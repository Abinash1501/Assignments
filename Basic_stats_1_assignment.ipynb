```python
# 1. Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
# 2. Load the Dataset
df = pd.read_csv(r"D:\Excelr\Assignments\Basic stats - 1\sales_data_with_discounts.csv")
df
```

```python
# 4. Descriptive Statistics for Numerical Columns
numerical_cols = ['Avg Price', 'Total Sales Value', 'Discount Rate (%)', 'Discount Amount', 'Net Sales Value']
desc_stats = df[numerical_cols].describe()
print("\nDescriptive Statistics:\n", desc_stats)
```

```python
# 5. Visualizations

# Histograms
for col in numerical_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], kde=True, bins=20)
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
# Boxplots
for col in numerical_cols:
    plt.figure(figsize=(6, 4))
    sns.boxplot(data=df, y=col)
    plt.title(f'Boxplot of {col}')
    plt.ylabel(col)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
# Bar Chart Analysis for Categorical Columns

categorical_cols = ['Day', 'SKU', 'City', 'BU', 'Brand', 'Model']

for col in categorical_cols:
    plt.figure(figsize=(10, 4))
    sns.countplot(data=df, x=col, order=df[col].value_counts().index)
    plt.title(f'Bar Chart of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.grid(True, axis='y', linestyle='--', linewidth=0.5)
    plt.show()
```

```python
# 6. Standardization of Numerical Variables
scaler = StandardScaler()
df_standardized = df.copy()
df_standardized[numerical_cols] = scaler.fit_transform(df[numerical_cols])
```

```python
print("\nStandardized Data (first 5 rows):")
print(df_standardized[numerical_cols].head())
```

```python
# 7. One-hot Encoding for Categorical Columns
categorical_cols = ['Day', 'SKU', 'City', 'BU', 'Brand', 'Model']
df_encoded = pd.get_dummies(df_standardized, columns=categorical_cols, drop_first=True)
```

```python
print("\nEncoded Data (first 5 rows):")
print(df_encoded.head())
```

```python
# One-hot encoding for categorical variables
categorical_columns = ['Day', 'SKU', 'City', 'BU', 'Brand', 'Model']
df_dummies = pd.get_dummies(df, columns=categorical_columns, drop_first=True)
df_dummies.head()
```

**Conclusion:**

**Key Findings:**

Descriptive analytics revealed the central tendencies (mean, median) and variability (standard deviation) in the dataset's numerical features, such as Avg Price, Total Sales, Discount %, and Net Sales Value.

Data visualizations like histograms and boxplots highlighted distributions, skewness, and the presence of outliers.

**Importance of Preprocessing:**

Standardization ensures that numerical features are on the same scale, which is critical for distance-based models and gradient-based optimizations in machine learning.

One-hot encoding transforms categorical variables into a machine-readable format, enabling algorithms to effectively process and learn from non-numerical data.
